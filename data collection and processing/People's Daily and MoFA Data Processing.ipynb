```python
import pandas as pd
```

## Processing People's Daily Data

```python
rmrb_data = pd.read_excel('PeoplesDailyData.xlsx')
```

```python
rmrb_data.rename(columns = {'article_headline':'headline', 'article_date':'date',
                              'article_text':'text'}, inplace = True)
rmrb_data.head()
```

```python
rmrb_usa = rmrb_data[rmrb_data['headline'].str.contains("U.S.|America|United States|Washington", regex = True).fillna(False)]
rmrb_usa = rmrb_usa.reset_index()
rmrb_usa.head()
```

```python
len(rmrb_usa)
# rmrb_usa.to_excel("PeoplesDailyUSA.xlsx")
```

## Processing MFA Press Conference Transcripts

```python
mfa = pd.read_excel("CMFA_PressCon_V2.xlsx")
```

```python
mfa.head()
# len(mfa)
```

```python
keywords_us = 'US|America|United States|Washington'
mfa_us = mfa[mfa.apply(lambda row: row.astype(str).str.contains(keywords_us, case=True).any(), axis=1)]
```

```python
mfa_us.head()
```

```python
mfa_us = mfa_us.drop(['id_orig','id_split','day','month','year'], axis = 1)
```

```python
# len(mfa_us)
mfa_us.to_excel("MFA_US.xlsx")
```
